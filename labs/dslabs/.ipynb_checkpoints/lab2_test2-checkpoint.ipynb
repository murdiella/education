```python
df = pd.read_csv('test2.csv')
df.head()
```

```python
lr_model = pickle.load(open('lr_model.pkl', 'rb'))
dt_model = pickle.load(open('dt_model.pkl', 'rb'))
rf_model = pickle.load(open('rf_model.pkl', 'rb'))
gb_model = pickle.load(open('gb_model.pkl', 'rb'))
```

```python
lifesquare_med = df['LifeSquare'].median()
df['LifeSquare'] = df['LifeSquare'].fillna(lifesquare_med)
Healthcare_1_med = df['Healthcare_1'].median()
df['Healthcare_1'] = df['Healthcare_1'].fillna(Healthcare_1_med)
missing_num = df.isnull().sum()
print(missing_num)
```

```python
lr_pred = lr_model.predict(df.drop(columns=['DistrictId', 'Ecology_2', 'Ecology_3', 'Id', 'Shops_2', 'Social_2'], axis=1))
lr = pd.DataFrame(lr_pred, columns=['predictions'])
lr.insert(0, 'Id', df['Id'])
lr.to_csv('lr_pred2.csv', index=False)
```

```python
dt_pred = dt_model.predict(df.drop(columns=['DistrictId', 'Ecology_2', 'Ecology_3', 'Id', 'Shops_2', 'Social_2'], axis=1))
dt = pd.DataFrame(dt_pred, columns=['predictions'])
dt.insert(0, 'Id', df['Id'])
dt.to_csv('dt_pred2.csv', index=False)
```

```python
rf_pred = rf_model.predict(df.drop(columns=['DistrictId', 'Ecology_2', 'Ecology_3', 'Id', 'Shops_2', 'Social_2'], axis=1))
rf = pd.DataFrame(rf_pred, columns=['predictions'])
rf.insert(0, 'Id', df['Id'])
rf.to_csv('rf_pred2.csv', index=False)
```
